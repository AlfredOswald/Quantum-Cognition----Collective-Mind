```python
# Installation von QuTiP (falls nicht bereits installiert)
# !pip install qutip

# Importieren der Bibliotheken
import numpy as np
import qutip as qt
from itertools import permutations
import matplotlib.pyplot as plt  # Für die Visualisierung
%matplotlib inline

# Basiszustände |0> und |1>
ket_0 = qt.basis(2, 0)  # Zustand |0>
ket_1 = qt.basis(2, 1)  # Zustand |1>

# Funktion zur Erstellung von Superpositionszuständen
def superposition_state(alpha, beta):
    norm = np.sqrt(abs(alpha)**2 + abs(beta)**2)
    return (alpha / norm) * ket_0 + (beta / norm) * ket_1

# Anzahl der Merkmale pro Person
num_traits = 2  # Offenheit und Gewissenhaftigkeit

# Qubit-Indizes
total_qubits = num_traits * 3  # 2 Merkmale * 3 Mitglieder

qubits_alice = list(range(0, num_traits))
qubits_bob = list(range(num_traits, num_traits * 2))
qubits_charlie = list(range(num_traits * 2, num_traits * 3))

# Szenarien mit unterschiedlichen Anfangszuständen der Gewissenhaftigkeit
scenarios = [
    {
        'name': 'Szenario 1 Gewissenhaftigkeit: Alice unbestimmt, Bob hoch, Charlie niedrig',
        'phi_C_Alice': superposition_state(1, 1),  # Alice unbestimmt
        'phi_C_Bob': ket_1,  # Bob hoch
        'phi_C_Charlie': ket_0  # Charlie niedrig
    },
    {
        'name': 'Szenario 2 Gewissenhaftigkeit: Alice hoch, Bob unbestimmt, Charlie niedrig',
        'phi_C_Alice': ket_1,  # Alice hoch
        'phi_C_Bob': superposition_state(1, 1),  # Bob unbestimmt
        'phi_C_Charlie': ket_0  # Charlie niedrig
    },
    {
        'name': 'Szenario 3 Gewissenhaftigkeit: Alice niedrig, Bob hoch, Charlie unbestimmt',
        'phi_C_Alice': ket_0,  # Alice niedrig
        'phi_C_Bob': ket_1,  # Bob hoch
        'phi_C_Charlie': superposition_state(1, 1)  # Charlie unbestimmt
    },
    {
        'name': 'Szenario 4 Gewissenhaftigkeit: Alle unbestimmt in Gewissenhaftigkeit',
        'phi_C_Alice': superposition_state(1, 1),  # Alice unbestimmt
        'phi_C_Bob': superposition_state(1, 1),  # Bob unbestimmt
        'phi_C_Charlie': superposition_state(1, 1)  # Charlie unbestimmt
    }
]

# Funktion zur Erstellung des Interaktionsoperators
def interaction_operator(phi_i_list, phi_j_list, qubits_i, qubits_j):
    # Start mit Nulloperator für den gesamten Hilbertraum
    interaction_op = qt.qzero([2]*total_qubits)

    for idx, (phi_i, phi_j) in enumerate(zip(phi_i_list, phi_j_list)):
        # Konstruktion des gemeinsamen Zustands für das Merkmal
        psi_ij = qt.tensor(phi_i, phi_j)
        psi_ji = qt.tensor(phi_j, phi_i)

        # Konstruktion des Ähnlichkeitsoperators für das Merkmal
        S_X = qt.ket2dm(psi_ij) + qt.ket2dm(psi_ji)

        # Erweiterung von S_X auf den gesamten Hilbertraum
        S_X_full = qt.expand_operator(S_X, dims=[2]*total_qubits, targets=[qubits_i[idx], qubits_j[idx]])

        # Hinzufügen zum gesamten Interaktionsoperator
        interaction_op += S_X_full

    return interaction_op

# Funktion zur Anwendung der Interaktionen in gegebener Reihenfolge
def apply_interactions(Psi_initial, interaction_sequence):
    Psi = Psi_initial
    for I in interaction_sequence:
        Psi = I * Psi
        Psi = Psi.unit()  # Normalisierung
    return Psi

# Funktion zur Berechnung des Zustandsindex
def state_index(qubit_states):
    index = 0
    for i, q_state in enumerate(reversed(qubit_states)):
        index += q_state * (2 ** i)
    return index

# Funktion zur Berechnung der Wahrscheinlichkeit für hohe Offenheit
def calculate_probability_high_openness(probabilities):
    # Berechnung der Gesamtwahrscheinlichkeit für hohe Offenheit bei allen Mitgliedern
    probability_sum = 0
    for gc_alice in [0, 1]:
        for gc_bob in [0, 1]:
            for gc_charlie in [0, 1]:
                qubit_states = [
                    1, gc_alice,  # Alice's Qubits: Offenheit hoch, Gewissenhaftigkeit variiert
                    1, gc_bob,    # Bob's Qubits: Offenheit hoch, Gewissenhaftigkeit variiert
                    1, gc_charlie # Charlie's Qubits: Offenheit hoch, Gewissenhaftigkeit variiert
                ]
                state_idx = state_index(qubit_states)
                prob_state = probabilities[state_idx]
                probability_sum += prob_state
    return probability_sum

# Funktion zur Berechnung der Wahrscheinlichkeit für hohe Gewissenhaftigkeit
def calculate_probability_high_conscientiousness(probabilities):
    # Berechnung der Gesamtwahrscheinlichkeit für hohe Gewissenhaftigkeit bei allen Mitgliedern
    probability_sum = 0
    for go_alice in [0, 1]:
        for go_bob in [0, 1]:
            for go_charlie in [0, 1]:
                qubit_states = [
                    go_alice, 1,  # Alice's Qubits: Offenheit variiert, Gewissenhaftigkeit hoch
                    go_bob, 1,    # Bob's Qubits: Offenheit variiert, Gewissenhaftigkeit hoch
                    go_charlie, 1 # Charlie's Qubits: Offenheit variiert, Gewissenhaftigkeit hoch
                ]
                state_idx = state_index(qubit_states)
                prob_state = probabilities[state_idx]
                probability_sum += prob_state
    return probability_sum

# Funktion zur Berechnung der Wahrscheinlichkeit für den "Collective Mind"-Zustand (Variante 1)
def calculate_probability_collective_mind_variant1(Psi_final):
    # Erstellung des "Collective Mind"-Zustands für Variante 1 (alle Qubits verschränkt)
    # Zustände |Φ_0⟩ und |Φ_1⟩
    Phi_0 = qt.tensor(
        ket_0, ket_0,  # Alice's Qubits
        ket_0, ket_0,  # Bob's Qubits
        ket_0, ket_0   # Charlie's Qubits
    )

    Phi_1 = qt.tensor(
        ket_1, ket_1,  # Alice's Qubits
        ket_1, ket_1,  # Bob's Qubits
        ket_1, ket_1   # Charlie's Qubits
    )

    Psi_CM = (Phi_0 + Phi_1).unit()  # Normierter "Collective Mind"-Zustand

    # Berechnung der Wahrscheinlichkeit
    prob_CM = np.abs(Psi_CM.overlap(Psi_final))**2
    return prob_CM

# Funktion zur Berechnung der Wahrscheinlichkeit für den "Collective Mind"-Zustand (Variante 2)
def calculate_probability_collective_mind_variant2(Psi_final):
    # Erstellung des "Collective Mind"-Zustands für Variante 2 (nur Offenheits-Qubits verschränkt)
    # Zustände |Ψ_0⟩ und |Ψ_1⟩ für Offenheits-Qubits
    Psi_O_0 = qt.tensor(
        ket_0,  # Alice's Offenheit
        ket_0,  # Bob's Offenheit
        ket_0   # Charlie's Offenheit
    )

    Psi_O_1 = qt.tensor(
        ket_1,  # Alice's Offenheit
        ket_1,  # Bob's Offenheit
        ket_1   # Charlie's Offenheit
    )

    Psi_O_CM = (Psi_O_0 + Psi_O_1).unit()  # Normierter "Collective Mind"-Zustand für Offenheit

    # Zustände der Gewissenhaftigkeits-Qubits (beliebiger Zustand)
    # Wir müssen über alle möglichen Zustände der Gewissenhaftigkeits-Qubits summieren
    prob_CM = 0
    for gc_alice in [0, 1]:
        for gc_bob in [0, 1]:
            for gc_charlie in [0, 1]:
                # Gewissenhaftigkeitszustände
                Psi_C = qt.tensor(
                    ket_0 if gc_alice == 0 else ket_1,  # Alice's Gewissenhaftigkeit
                    ket_0 if gc_bob == 0 else ket_1,    # Bob's Gewissenhaftigkeit
                    ket_0 if gc_charlie == 0 else ket_1 # Charlie's Gewissenhaftigkeit
                )

                # Gesamtzustand: Offenheits-"Collective Mind"-Zustand und spezifische Gewissenhaftigkeitszustände
                Psi_CM_full = qt.tensor(Psi_O_CM, Psi_C)

                # Berechnung des Überlapps mit dem finalen Zustand
                overlap = np.abs(Psi_CM_full.overlap(Psi_final))**2
                prob_CM += overlap

    return prob_CM

# Hauptprogramm
# Generiere alle möglichen Permutationen der Interaktionsoperatoren
operator_names = ['I_12', 'I_23', 'I_13']
interaction_sequences = list(permutations(operator_names))

# Initialisierung der Gesamtergebnisse
all_results = []

for scenario in scenarios:
    print(f"\n=== {scenario['name']} ===")
    # Alice's Zustände
    phi_O_Alice = ket_1  # Offenheit hoch
    phi_C_Alice = scenario['phi_C_Alice']
    psi_Alice = qt.tensor(phi_O_Alice, phi_C_Alice)

    # Bob's Zustände
    phi_O_Bob = superposition_state(1, 1)  # Offenheit unbestimmt
    phi_C_Bob = scenario['phi_C_Bob']
    psi_Bob = qt.tensor(phi_O_Bob, phi_C_Bob)

    # Charlie's Zustände
    phi_O_Charlie = superposition_state(1, 1)  # Offenheit unbestimmt
    phi_C_Charlie = scenario['phi_C_Charlie']
    psi_Charlie = qt.tensor(phi_O_Charlie, phi_C_Charlie)

    # Gemeinsamer Zustand des Teams
    Psi_initial = qt.tensor(psi_Alice, psi_Bob, psi_Charlie)

    # Erstellung der Listen der Zustände
    phi_Alice_list = [phi_O_Alice, phi_C_Alice]
    phi_Bob_list = [phi_O_Bob, phi_C_Bob]
    phi_Charlie_list = [phi_O_Charlie, phi_C_Charlie]

    # Erstellung der Interaktionsoperatoren
    I_12 = interaction_operator(phi_Alice_list, phi_Bob_list, qubits_alice, qubits_bob)
    I_23 = interaction_operator(phi_Bob_list, phi_Charlie_list, qubits_bob, qubits_charlie)
    I_13 = interaction_operator(phi_Alice_list, phi_Charlie_list, qubits_alice, qubits_charlie)

    # Liste der Interaktionsoperatoren
    interaction_operators = [I_12, I_23, I_13]

    # Erstelle alle möglichen Reihenfolgen der Interaktionsoperatoren
    operator_permutations = list(permutations(interaction_operators))
    sequence_names = list(permutations(operator_names))

    # Initialisierung der Ergebnisse
    results = []

    # Durchlaufen aller möglichen Reihenfolgen
    for seq_ops, seq_names in zip(operator_permutations, sequence_names):
        # Anwenden der Interaktionen
        Psi_final = apply_interactions(Psi_initial, seq_ops)

        # Berechnung der Wahrscheinlichkeiten
        probabilities = np.abs(Psi_final.full())**2
        probabilities = probabilities.flatten()

        # Wahrscheinlichkeit für hohe Offenheit bei allen Mitgliedern
        prob_openness = calculate_probability_high_openness(probabilities)

        # Wahrscheinlichkeit für hohe Gewissenhaftigkeit bei allen Mitgliedern
        prob_conscientiousness = calculate_probability_high_conscientiousness(probabilities)

        # Wahrscheinlichkeit für "Collective Mind" Variante 1
        prob_CM_variant1 = calculate_probability_collective_mind_variant1(Psi_final)

        # Wahrscheinlichkeit für "Collective Mind" Variante 2
        prob_CM_variant2 = calculate_probability_collective_mind_variant2(Psi_final)

        # Speichere das Ergebnis
        results.append({
            'sequence': ' → '.join(seq_names),
            'prob_openness': prob_openness,
            'prob_conscientiousness': prob_conscientiousness,
            'prob_CM_variant1': prob_CM_variant1,
            'prob_CM_variant2': prob_CM_variant2
        })

    # Ausgabe der Ergebnisse für das Szenario
    print("Ergebnisse für verschiedene Reihenfolgen der Interaktionen:")
    for res in results:
        print(f"Reihenfolge: {res['sequence']}, "
              f"Wahrscheinlichkeit Offenheit: {res['prob_openness']:.4f}, "
              f"Wahrscheinlichkeit Gewissenhaftigkeit: {res['prob_conscientiousness']:.4f}, "
              f"Wahrscheinlichkeit Collective Mind Variante 1: {res['prob_CM_variant1']:.4f}, "
              f"Wahrscheinlichkeit Collective Mind Variante 2: {res['prob_CM_variant2']:.4f}")

    # Speichere die Ergebnisse für alle Szenarien
    all_results.append({
        'scenario': scenario['name'],
        'results': results
    })

    # Visualisierung der Ergebnisse für das aktuelle Szenario
    # Sortieren der Ergebnisse nach Wahrscheinlichkeit für Offenheit
    results_sorted = sorted(results, key=lambda x: x['prob_openness'], reverse=True)

    # Extrahieren der Sequenzen und Wahrscheinlichkeiten
    sequences = [res['sequence'] for res in results_sorted]
    prob_openness = [res['prob_openness'] for res in results_sorted]
    prob_conscientiousness = [res['prob_conscientiousness'] for res in results_sorted]
    prob_CM_variant1 = [res['prob_CM_variant1'] for res in results_sorted]
    prob_CM_variant2 = [res['prob_CM_variant2'] for res in results_sorted]

    # Erstellen des Balkendiagramms
    x = np.arange(len(sequences))  # Positionen auf der x-Achse
    width = 0.15  # Breite der Balken

    plt.figure(figsize=(12, 6))
    bar1 = plt.bar(x - 1.5*width, prob_openness, width, label='Hohe Offenheit', color='skyblue')
    bar2 = plt.bar(x - 0.5*width, prob_conscientiousness, width, label='Hohe Gewissenhaftigkeit', color='salmon')
    bar3 = plt.bar(x + 0.5*width, prob_CM_variant1, width, label='Collective Mind Variante 1', color='green')
    bar4 = plt.bar(x + 1.5*width, prob_CM_variant2, width, label='Collective Mind Variante 2', color='purple')

    plt.xlabel('Reihenfolge der Interaktionen')
    plt.ylabel('Wahrscheinlichkeit')
    plt.title(f"Einfluss der Interaktionsreihenfolge auf Eigenschaften\n{scenario['name']}")
    plt.xticks(x, sequences, rotation=45, ha='right')
    plt.legend()

    # Werte über den Balken anzeigen
    for bars in [bar1, bar2, bar3, bar4]:
        for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.005, f'{yval:.4f}', ha='center', va='bottom')

    plt.tight_layout()
    plt.show()
```
